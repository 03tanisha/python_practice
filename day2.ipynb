```python
# bank system 
class Bankaccount:
    def __init__(self, acc_holder, acc_num, balance):
        self.acc_holder = acc_holder 
        self.acc_num = acc_num 
        self.__balance = balance 


    # deposit method 
    def deposit(self, amount):
        if amount > 0 :
            self.__balance += amount
            print("amount has been deposited successfully")
        else:
            print("insufficient amount to deposit")

    # withdraw method 
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount 
            print("amount has been withdrawn successfully")

        else:
            print("insufficient balance to withdraw")

    # get balnce method
    def get_balance(self):
        return self.__balance
    
acc1 = Bankaccount("bob", "12345", 1000)
acc2 = Bankaccount("tanisha", "6789", 2000)
```

```python
# calling the methods 
acc1.deposit(5)
acc1.withdraw(1000)
print(acc1.get_balance())
print()
acc2.deposit(7000)
acc2.withdraw(200)
print(acc2.get_balance())
```

```python
# student mark system 
class Student:
    def __init__(self, name, course, marks):
        self.name = name
        self.course = course
        self.__marks = marks 

    # set marks method 
    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks 
        else:
            print("fail")

    # get marks method 
    def get_marks(self):
        return self.__marks
    
# objects 
s1 = Student("tanisha","aieng",100)
s2 = Student("sameer", "cs", 40)
```

```python
# calling the methods
s1.set_marks(120)
print(s1.get_marks())
print()
s2.set_marks(85)
print(s2.get_marks())
```

```python
class Thermostat:
    def __init__(self, place, temp=24):
        self.place = place
        self.__temp = temp

    # increase temp method 
    def inc_temp(self, value):
        if self.__temp + value <= 30:
            self.__temp += value 
        else:
            self.__temp = 30
            print("maximum temp. reached")

    # decrease temp method
    def dec_temp(self, value):
        if self.__temp - value >=16:
            self.__temp -= value
        else:
            self.__temp = 16
            print("minimum temp. reached")
       
    # get temp method 
    def get_temp(self):
        return self.__temp
    
t1 = Thermostat("delhi")
```

```python
# calling methods 
t1.inc_temp(5)
print(t1.get_temp())

t1.inc_temp(40)
print(t1.get_temp())

t1.dec_temp(15)
print(t1.get_temp())

t1.dec_temp(5)
print(t1.get_temp())
```
