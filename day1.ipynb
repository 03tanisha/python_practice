```python
# student managemnt system 
 
class Student:
    def __init__(self, name, course, roll_num,marks):
        self.name = name 
        self.course = course
        self.roll_num = roll_num
        self.marks = marks 


    def display_details(self):
        print(f"name :",self.name)
        print(f"course:",self.course)
        print(f"roll_num:",self.roll_num)
        print(f"marks:",self.marks)

    def check_result(self):
        if self.marks >= 40:
            print("pass")
        else:
            print("fail")


student1 = Student("ram", "btech", 101, 85)
student2 = Student("tanisha", "btech",102,34)
```

```python
# calling the methods 

student1.display_details()
student1.check_result()
print()
student2.display_details()
student2.check_result()
```

```python
# bank system 

class BankAccount:
    def __init__(self, acc_holder, acc_num, balance):
        self.acc_holder = acc_holder
        self.acc_num = acc_num
        self.balance = balance 

    def display_acc_details(self):
        print(f"account holder:", self.acc_holder)
        print(f"account number:", self.acc_num)
        print(f"balance:", self.balance)

    def deposit(self, amount):
        self.balance += amount
        print(f"deposited {amount} successfully. New balance is {self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount 
            print(f"withdrew {amount} successfully. New balance is {self.balance}")
        else:
            print("insufficient balance")

acc1 = BankAccount("tanisha", 12345, 5000)
acc2 = BankAccount("tanu",6789, 2000)
```

```python
# calling methods 
acc1.display_acc_details()
acc1.deposit(1500)
acc1.withdraw(7500)
print()
acc2.display_acc_details()
acc2.deposit(4000)
acc2.withdraw(200)
```

```python
# car system 

class Car:
    def __init__(self, brand,model,year, speed):
        self.brand = brand 
        self.model = model
        self.year = year
        self.speed = speed

    def start(self):
        print("starting the car")

    def accelerate(self, speed):
        self.speed += speed
        print(f"car accelerated to {self.speed}km/h")

    def show_details(self):
        print(f"brand:", self.brand)
        print(f"model:", self.model)
        print(f"year:",self.year)

car1 = Car("maurti suzuki","swift", 2024, 120)
car2 = Car("mahindra","thar", 2025, 150)
```

```python
# calling methods
car1.show_details()
car1.start()
car1.accelerate(30)
car1.show_details()
print()
car2.show_details()
car2.start()
car2.accelerate(50)
car2.show_details()
```
